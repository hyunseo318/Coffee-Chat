
##데이터전처리

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/커피챗_11월/커피챗_데이터.csv')
df.head()
```

```python
df.info()
```

```python
real_crew = [
    '제이시', '풀리', '루나','엘리', '쏠', '리건', '올리', '라쿤', '바이올렛', '제이미', '셀린',
    '릴리', '리지', '투나', '코바', '레이첼','아르코','브리애나','버즈','토니','빈센트','댄','애나','켈리','다나',
    '데이빗','얀','데인','벨','샘'
]
```

```python
len(real_crew)
```

```python
# contents 특수문자 제거
#df['contents']= df['contents'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)
df.head()
```

```python
for i in range(len(df['contents'])):
  df['contents'][i]=df['contents'].str.split('\n')[i]
  del df['contents'][i][0]
  for word in df['contents'][i]:
    if word.startswith('@'):
      df['contents'][i].remove(word)
df.head()
```

```python
df['crew']=df['crew'].str.split(',').tolist()
for i in range(len(df['crew'])):
  for j in range(len(df['crew'][i])):
    for r in df['crew'][i]:
      if r not in real_crew:
        df['crew'][i].remove(r)

df
```

```python
now = df.join(df['crew'].apply(lambda x: pd.Series(x)).stack().reset_index(1,name='crew_sep').drop('level_1', axis=1))
now
```

```python
now2 = now[['crew_sep', 'contents']]
now2.reset_index(inplace=True, drop=True)
now2
```

```python
now_exist=now2.dropna()
now_exist
```

```python
now_exist.reset_index(drop=True)
```

```python
idx = now_exist[now_exist['contents'] == '-1'].index
now_exist = now_exist.drop(idx)
now_exist
```

```python
result = pd.DataFrame(real_crew, columns=['crew_sep'])

result
```

```python
result2=pd.merge(result,now_exist)
```

```python
result2.loc[2]['contents']
```

```python
' '.join(result2.loc[2]['contents'])
```

```python
result2['contents_ver2'] = result2['contents'].apply(lambda x: ' '.join(x))
```

```python
result2
```

```python
result2['text'] = result2.groupby(['crew_sep'])[['contents_ver2']].transform(lambda x: ' '.join(x))
final=result2[['crew_sep','text']].drop_duplicates().reset_index(drop=True)
final
```

```python
final["text"] = final["text"].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
```

```python
for i in range(len(final)):
  rr = ' '.join(str(final['text'][i]).split())
  final['text'][i] = rr
final
```

```python
crew_name = []
for i in list:
  temp = i.split('/')[6]
  temp = temp.split('.')[0]
  crew_name.append(temp)

crew_name
```

```python
# 벡터 가져오는 목적

from transformers import ViTFeatureExtractor, ViTModel

feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')
model = ViTModel.from_pretrained('google/vit-base-patch16-224-in21k')
```

```python
import torch
from PIL import Image

result1 = pd.DataFrame()

for img in list:
  image = Image.open(img) 
  inputs = feature_extractor(images=image, return_tensors="pt")
  outputs = model(**inputs)
  vec = outputs.pooler_output

  temp = pd.DataFrame(vec.detach().numpy())
  result1 = pd.concat([result1, temp])

result1
```

```python
result1.index = crew_name
result1
```
